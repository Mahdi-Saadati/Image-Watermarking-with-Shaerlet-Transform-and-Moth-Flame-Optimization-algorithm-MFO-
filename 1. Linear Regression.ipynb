هرچی بیشتر درس بخونی، نمره‌ات چقدر بهتر میشه؟ 🎯

مثلاً چندتا مثال داری:

اگر این نقاط رو روی یک کاغذ رسم کنی، تقریباً روی یک خط می‌افتند.
کامپیوتر می‌آید بهترین خط ممکن را بین این نقطه‌ها پیدا می‌کند.
بعد می‌تواند حدس بزند:
اگر ۵ ساعت درس بخونی، تقریباً چند می‌گیری؟ 🎉

به این روش حدس‌زدن می‌گویند: رگرسیون خطی یعنی «یک خط پیدا کن که داده‌ها رو بهترین شکل ممکن توضیح بده.»

✨ خیلی ساده:
📈 ورودی: ساعت درس خواندن

📈 خروجی: نمره

📈 کامپیوتر: یک خط می‌کشه که به داده‌ها بیشترین شباهت رو داشته باشه.

بعد می‌گه: اگه ورودی جدید بدی (مثلاً ۵ ساعت)، روی اون خط نگاه می‌کنه و جواب رو حدس می‌زنه.

✅ یک مثال جذاب پایتون
حالا بیایم همون داستان درس‌خوندن و نمره رو با پایتون و یک نمودار خوشگل ببینیم! 🎨
</div>



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# داده‌ها: ساعت مطالعه و نمره
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
scores = np.array([50, 60, 70, 80, 90])

# ساخت مدل و آموزش
model = LinearRegression()
model.fit(hours, scores)

# پیش‌بینی برای یک ساعت جدید (مثلا 6 ساعت)
new_hour = np.array([[6]])
predicted_score = model.predict(new_hour)
print(f"اگر 6 ساعت درس بخونی نمره‌ات حدودا میشه: {predicted_score[0]:.1f}")

# رسم نمودار
plt.scatter(hours, scores, color='blue', label='داده‌های واقعی 📚')
plt.plot(hours, model.predict(hours), color='red', label='خط یادگرفته‌شده 🔥')
plt.scatter(new_hour, predicted_score, color='green', s=100, label='حدس جدید 🎯')
plt.xlabel('ساعت مطالعه')
plt.ylabel('نمره')
plt.title('پیش‌بینی نمره با رگرسیون خطی')
plt.legend()
plt.grid(True)
plt.show()
```
