```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ÿØÿßÿØŸá‚ÄåŸáÿß: ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá Ÿà ŸÜŸÖÿ±Ÿá
hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
scores = np.array([50, 60, 70, 80, 90])

# ÿ≥ÿßÿÆÿ™ ŸÖÿØŸÑ Ÿà ÿ¢ŸÖŸàÿ≤ÿ¥
model = LinearRegression()
model.fit(hours, scores)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿ®ÿ±ÿß€å €å⁄© ÿ≥ÿßÿπÿ™ ÿ¨ÿØ€åÿØ (ŸÖÿ´ŸÑÿß 6 ÿ≥ÿßÿπÿ™)
new_hour = np.array([[6]])
predicted_score = model.predict(new_hour)
print(f"ÿß⁄Øÿ± 6 ÿ≥ÿßÿπÿ™ ÿØÿ±ÿ≥ ÿ®ÿÆŸàŸÜ€å ŸÜŸÖÿ±Ÿá‚Äåÿßÿ™ ÿ≠ÿØŸàÿØÿß ŸÖ€åÿ¥Ÿá: {predicted_score[0]:.1f}")

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.scatter(hours, scores, color='blue', label='ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸàÿßŸÇÿπ€å üìö')
plt.plot(hours, model.predict(hours), color='red', label='ÿÆÿ∑ €åÿßÿØ⁄Øÿ±ŸÅÿ™Ÿá‚Äåÿ¥ÿØŸá üî•')
plt.scatter(new_hour, predicted_score, color='green', s=100, label='ÿ≠ÿØÿ≥ ÿ¨ÿØ€åÿØ üéØ')
plt.xlabel('ÿ≥ÿßÿπÿ™ ŸÖÿ∑ÿßŸÑÿπŸá')
plt.ylabel('ŸÜŸÖÿ±Ÿá')
plt.title('Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ŸÜŸÖÿ±Ÿá ÿ®ÿß ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ÿÆÿ∑€å')
plt.legend()
plt.grid(True)
plt.show()
```
